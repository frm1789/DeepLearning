```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the function f(x)
def f(x):
    return x ** 2
```

```python
# Plot the function
x_vals = np.linspace(-5, 5, 100)  # Create an array of x values for plotting
y_vals = f(x_vals)  # Compute the corresponding y values for the function

plt.plot(x_vals, y_vals, label='f(x) = x^2')  # Plot the function

# Highlight the point where x = -1.5
plt.scatter(-1.5, f(-1.5), color='red', label="Point (-1.5, f(-1.5))")

plt.legend()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of f(x) = x^2 with a red point at x = -1.5')

# Show the plot
plt.show()
```

```python
xt = tensor(3.).requires_grad_()
```

```python
yt = f(xt)
yt
```
